## Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split
from classification import ML_Classifier, load_model
```

## Load Dataset

```python
data = pd.read_csv("heart.csv")
data.head()
```

## Preprocessing

```python
X = data.drop("output",axis=1)
Y = data.output.values
```

## Sclaing:
### StandardScaler

```python
SS = StandardScaler()
X = SS.fit_transform(X)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

## ML_Classifier
### ML_Classifier include 8 models:
* KNN
* SVC
* DecisionTree
* RandomForest
* GBoost
* AdaBoost
* XGBoost
* CatBoost

### ML_Classifier include 4 scores:
* Accuracy
* Precision
* Recall
* F1

### This model finds the best model based on the score chosen by the user

```python
model = ML_Classifier(score="accuracy")
```

```python
model.fit_data(x_train,y_train,x_test,y_test)
```

```python
predicted = model.predict_data(x_test)
```

### Save Model

```python
model.save_model()
```
